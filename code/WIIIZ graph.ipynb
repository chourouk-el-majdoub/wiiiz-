```python
import sqlalchemy as sql
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
from jinja2 import Environment, FileSystemLoader
engine = create_engine("mysql+pymysql://root:Simplon1:@localhost/WIIIZ")

sqlQuery = pd.read_sql_query("select * FROM Borne join Enregistrement on BO_ID = Enregistrement.EN_borne_id_FK JOIN Utilisation on UT_ID = Enregistrement.EN_utilisation_id_FK LIMIT 3000", engine)
                                                                                       
dfwiiiz = pd.DataFrame(sqlQuery)

dfwiiiz
```

```python
listemois=['janvier', 'fevrier', 'mars', 'avril', 'mai', 'juin', 'juillet', 'aout','septembre','octobre','novembre','decembre']
for i in dfwiiiz.index:
    dfwiiiz.loc[i, "mois"] = listemois[dfwiiiz.loc[i, "UT_Mois_fin"]-1]

plt.figure(figsize=[9,10])
dfwiiiz["mois"].value_counts(normalize=True).plot(kind='pie',autopct = lambda x: str(round(x, 2)) + '%')
plt.title("Nombre d'utilisations par mois")
plt.legend()
```

```python
listeVilles = ["Cannes", "Le Cannet", "Mandelieu", "Mougins", "Théoule"]
nbBornesVilles = [0] * len(listeVilles)

for borne in dfwiiiz["BO_nom"].unique():
    for i, val in enumerate(listeVilles):
        if val in borne or val.upper() in borne:
            nbBornesVilles[i] += 1

for i, val in enumerate(listeVilles):
    print(val + " possède " + str(nbBornesVilles[i]) + " bornes")
```

```python
listeVilles = ["Cannes", "Le Cannet", "Mandelieu", "Mougins", "Théoule"]

dfVilles = pd.DataFrame(columns = ['borne','ville'])

dfVilles['borne'] = dfwiiiz["BO_nom"].unique()

for i, borne in enumerate(dfVilles["borne"]):
    for val in listeVilles:
          if val in borne or val.upper() in borne:
                dfVilles.loc[i, 'ville'] = val

dfVilles
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=[10,10])
dfVilles['ville'].value_counts(normalize=True).plot(kind='pie', autopct = lambda x: str(round(x, 2)) + '%',explode = [0.1, 0, 0, 0, 0.1])

plt.title("Nombre de bornes par ville")

plt.legend()
```

```python
dfwiiiz["heureDebut"] = pd.to_datetime(dfwiiiz["UT_Date_debut"]).dt.hour

donneesparheure = dfwiiiz.groupby(by=["UT_Heure_debut", "BO_nom"]).size().reset_index(name="count")

plt.figure(figsize= [20, 16])
sns.pointplot(x="UT_Heure_debut", y="count", hue="BO_nom", data=donneesparheure)
plt.title("Nombre d'utilisation par heure et par Borne")
autopct = lambda x: str(round(x, 2)) + '%',
plt.legend()
```
