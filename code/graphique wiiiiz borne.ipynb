```python
import sqlalchemy as sql
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
from jinja2 import Environment, FileSystemLoader

engine=create_engine("mysql+pymysql://simplon:simplon@localhost:3306/wizz")
sqlquery=pd.read_sql_query("select * from Borne join Enregistrement on Enregistrement.BO_ID = Borne.BO_ID join Utilisation on Utilisation.UT_ID = Enregistrement.UT_ID", engine)

dfwiiiz = pd.DataFrame(sqlquery)
dfwiiiz
```

```python

plt.figure(figsize=[20,15])
plt.subplot(111)
sns.countplot(data=dfwiiiz , y='BO_nom')

plt.xticks(rotation=90)
```

```python
listemois=['janvier', 'fevrier', 'mars', 'avril', 'mai', 'juin', 'juillet', 'aout','septembre','octobre','novembre','decembre']
for i in dfwiiiz.index:    
    dfwiiiz.loc[i, "mois"] = listemois[dfwiiiz.loc[i, "UT_Mois_fin"]-1]
plt.figure(figsize=[7,10])
dfwiiiz["mois"].value_counts(normalize=True).plot(kind='pie',autopct = lambda x: str(round(x, 2)) + '%',explode = [0, 0.1, 0, 0, 0,0,0,0])
plt.title("Nombre d'utilisations par mois")
plt.legend()
```

```python
plt.figure(figsize=[7,10])
dfwiiiz["BO_fabricant"].value_counts(normalize=True).plot(kind='pie',autopct = lambda x: str(round(x, 2)) + '%')
plt.title("Nombre d'utilisations par fabricant")
plt.legend()
```

```python
dfwiiiz["BO_prise"].value_counts(normalize=True).plot(kind='pie',autopct = lambda x: str(round(x, 2)) + '%')
plt.title("Nombre d'utilisations par type de prise")
plt.legend()
```

```python
listeVilles = ["Cannes", "Le Cannet", "Mandelieu", "Mougins", "Théoule"]
nbBornesVilles = [0] * len(listeVilles)

for borne in dfwiiiz["BO_nom"].unique():
    for i, val in enumerate(listeVilles):
        if val in borne or val.upper() in borne:
            nbBornesVilles[i] += 1

for i, val in enumerate(listeVilles):
    print(val + " possède " + str(nbBornesVilles[i]) + " bornes")
```

```python
listeVilles = ["Cannes", "Le Cannet", "Mandelieu", "Mougins", "Théoule"]

dfVilles = pd.DataFrame(columns = ['borne','ville'])

dfVilles['borne'] = dfwiiiz["BO_nom"].unique()

for i, borne in enumerate(dfVilles["borne"]):
    for val in listeVilles:
          if val in borne or val.upper() in borne:
                dfVilles.loc[i, 'ville'] = val

dfVilles
```

```python
plt.figure(figsize=[10,10])
dfVilles['ville'].value_counts(normalize=True).plot(kind='pie',autopct = lambda x: str(round(x, 2)) + '%')

plt.title("Nombre de bornes par ville")

plt.legend()
```

```python
listeVilles = ["Cannes", "Le Cannet", "Mandelieu", "Mougins", "Théoule"]

for i in dfwiiiz.index:
    for val in listeVilles:
        if val in dfwiiiz.loc[i, "BO_nom"] or val.upper() in dfwiiiz.loc[i, "BO_nom"]:
            dfwiiiz.loc[i, 'ville'] = val
                
dfwiiiz
```

```python
donneesparheure = dfwiiiz.groupby(by=["UT_Heure_debut", "ville"]).size().reset_index(name="count")

plt.figure(figsize= [20, 16])
sns.pointplot(x="UT_Heure_debut", y="count", hue="ville", data=donneesparheure)
plt.title("Nombre d'utilisation par heure et par ville")
```

```python
donneesparmois = dfwiiiz.groupby(by=["UT_Mois_debut", "ville"]).size().reset_index(name="count")

plt.figure(figsize= [20, 16])
sns.pointplot(x="UT_Mois_debut", y="count", hue="ville", data=donneesparmois)
plt.title("Nombre d'utilisation par mois et par ville")
```
